```python
df['Description']=df['Description'].str.strip() #removeing spaces from description if have
df.dropna(subset=['InvoiceNo'],axis=0,inplace=True) # droping null values.
df['InvoiceNo']=df['InvoiceNo'].astype('str') # changing invoice no into str.
df=df[~df['InvoiceNo'].str.contains('C')] #Removeing credit card values
df.head()
```

```python
df.info()
```

```python
df['Country'].value_counts()
```

```python
#groping the data by using country is United kingdom
my_basket = (df[df['Country'] =="Germany"]
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))
```

```python
my_basket.head()
```

```python
# Encoding the data where value of x is 0 or smaller than 0,in return it gonna give me 0 value,
# like that for those values which is greter than 0 it will return 1.
def encoder(x):
    if x<=0:
        return 0
    if x>0:
        return 1
my_basket_sets=my_basket.applymap(encoder)
my_basket_sets.head()
```

```python
my_basket_sets['POSTAGE'].isnull().sum()
```

```python
del my_basket_sets['POSTAGE']
```

```python
my_basket_sets.head()
```

```python
#Here i used apriori in order to generate the frequency of item sets
frequent_itemsets=apriori(my_basket_sets,min_support=0.07,use_colnames=True)
```

```python
#Generating some rules
rules=association_rules(frequent_itemsets,metric='lift',min_threshold=1)
rules.head()
```

# Let's do some predictions

```python
rules[ (rules['lift'] >= 3) &
       (rules['confidence'] >= 0.5) ]
```
